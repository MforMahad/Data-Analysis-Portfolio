```python
#removing extra spaces
df.columns = df.columns.str.strip() 
```

```python
#removing space and adding "_" for making things easier in future

df.columns = [c.replace(' ', '_') for c in df.columns] 
```

```python
# now fixing Data type of Transaction_Date
df['Transaction_Date']  = pd.to_datetime(df['Transaction_Date'], errors = 'coerce')
```

```python
df.info()
```

```python
# Now Fixing The Numeric Columns

numeric_cols = ['Price_Per_Unit', 'Quantity', 'Total_Spent']
for col in numeric_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace('[$,]', '', regex = True)
    df[col] = pd.to_numeric(df[col], errors = 'coerce')
```

```python
#Can't delete those null item values because the next necessary columns are not checked yet so will name all nul items 
#as Unknown Item
df['Item'] = df['Item'].fillna('Unknown_Item')
```

```python
df['Item'].isnull().sum()
```

```python
# 1. Identify rows where BOTH Quantity and Total_Spent are missing
# We use .dropna with the 'subset' and 'how' parameters
df_cleaned = df.dropna(subset = ['Quantity', 'Total_Spent'], how = 'all')
```

```python
# now will see how many rows we actually removed
rows_removed = len(df) - len(df_cleaned)
print(f"Cleaned! We removed {rows_removed} rows that had no Quantity And Total_Spent.")
```

```python
# Check if there are rows where money is 0 but quantity exists
zero_sales = df[(df['Total_Spent'] == 0) & (df['Quantity'] > 0)]
print(f"Transactions with 0 cost but > 0 quantity: {len(zero_sales)}")
```

```python
df['Price_Per_Unit'].isnull().sum()
```

```python
# Check if any of those 609 rows have a Quantity of 0 or NaN
check_zeros = df[df['Price_Per_Unit'].isnull()]['Quantity'].value_counts(dropna=False)
print("Quantity values for the missing price rows:")
print(check_zeros)
```

```python
#Recovering missing prices from Total and Quantity
df.loc[:, 'Price_Per_Unit'] = df['Price_Per_Unit'].fillna(df['Total_Spent'] / df['Quantity'])
```

```python
#now it will give 0 null values
df['Price_Per_Unit'].isnull().sum()
```

```python
print(f"Missing Quantities: {df['Quantity'].isnull().sum()}")
```

```python
print(f"Missing Total: {df['Total_Spent'].isnull().sum()}")
```

```python
df['Discount_Applied'].isnull().sum()
```

```python
#Creating the 'Calculated' price
calculated_total = df['Price_Per_Unit'] * df['Quantity']
```

```python
# Check if there is ANY row where the calculation is different from the actual Total_Spent
mismatches = (df['Total_Spent'] - calculated_total).abs() > 0.01   
```

```python
# Count how many rows failed the test
mismatch_count = mismatches.sum()
```

```python
print(f"Number of rows where Price * Quantity != Total: {mismatch_count}")
```

```python
# If the count is 0, then we have PROOF to set everything to False
if mismatch_count == 0:
    df['Discount_Applied'] = False
```

```python
df['Item'] = df['Item'].replace('Unknown Item', 'Unknown_Item')
```

```python
print(df.dtypes)
```

```python
#now we can see those empty items are unknown items
#and will remove Discount Applied column as there are no DIscounts given
df.describe(include='all')
```

```python
#Convert Quantity to Integer (since there are no 0.5 quantities)
df['Quantity'] = df['Quantity'].astype(int)
```

```python
#To optimize the data for analysis and remove redundant features, I dropped this column.
df = df.drop(columns = ['Discount_Applied'])
```
