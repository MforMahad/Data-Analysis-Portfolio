```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
```

```python
#Creating a month Column to check when do customers spend the most

df['Month'] = df['Transaction_Date'].dt.month
```

```python
df.head()
```

```python
#Now we Will DO EDA on the Dateset we cleaned
# 1 Which category generates the most revenue

df.groupby('Category')['Total_Spent'].sum().sort_values(ascending = False).reset_index()
```

```python
# 2 Which payment method is used the most?

df.groupby('Payment_Method')['Transaction_ID'].count().reset_index()
```

```python
# 2 Which payment method is used the most?its same mostly professionals use this method
df["Payment_Method"].value_counts().reset_index()
```

```python
# 3 Whether online or physical store generates more revenue.

df.groupby('Location')['Total_Spent'].sum().sort_values(ascending = False).reset_index()
```

```python
# 4 Identify high-value customers.

df.groupby('Customer_ID')['Total_Spent'].sum().sort_values(ascending = False).reset_index().head()
```

```python
# 6 Which category sells the most items (not revenue)?

df.groupby('Category')['Quantity'].sum().sort_values(ascending = False).reset_index()
```

```python
# 5 Monthly Sales Trend

df.groupby('Month')['Total_Spent'].sum().sort_values(ascending = False).reset_index().head()
```
