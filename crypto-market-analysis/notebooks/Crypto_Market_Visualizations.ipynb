```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('C:/Users/Dell/Desktop/pythonDS/pythonfordsagain/Python API Collector/crypto_market_data_v3.csv')
```

```python
# Top 10 Coins by Market Cap (Bar Chart)

top_10_by_market_cap = df.sort_values('market_cap',ascending = False).head(10)
```

```python
plt.figure(figsize = (10,6))

sns.barplot(x = 'market_cap', y = 'name', data = top_10_by_market_cap)


plt.title('Top 10 By Market Cap', fontsize=14, weight='bold')
plt.xlabel('Market Cap')
plt.ylabel('Cryptocurrency')

plt.show()
```

```python
# Highest Trading Volume Coins

high_trad_vol_coins = df.sort_values('total_volume', ascending = False).head(10)
```

```python
plt.figure(figsize = (10,6))

sns.barplot(x = 'total_volume', y = 'name', data =high_trad_vol_coins)

plt.title('Highest Trading Volume Coins',fontsize=14, weight='bold')
plt.xlabel('Total Volume')
plt.ylabel('Cryptocurrency')

plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e9:.1f}B'))

plt.show()
```

```python
# Biggest 24h Gainers

top_gainer = df.sort_values('price_change_percentage_24h', ascending = False).head(10)
```

```python
plt.figure(figsize = (10, 6))

sns.barplot(x = 'price_change_percentage_24h', y = 'name', data = top_gainer)

plt.title('Biggest 24h Gainers', fontsize=14, weight='bold')
plt.xlabel('24h Price Change (%)')
plt.ylabel('Cryptocurreny')


plt.show()
```

```python
# Volatility Scatter Plot (Very Analyst-Like)

plt.figure(figsize = (10, 6))
sns.scatterplot(x = 'market_cap', y = 'volatility_estimate', data = df)

plt.title('Market Cap vs Volatility',fontsize=14, weight='bold')
plt.xlabel('Market Cap')
plt.ylabel('Volatility Estimate')

plt.xticks(rotation=45)

plt.show()
```
